```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

```python
bitcoin_data= cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency='usd',days=30)
```

```python
print(bitcoin_data)
```

```python
import pandas as pd
```

```python
bitcoin_price_data=bitcoin_data['prices']
```

```python
data=pd.DataFrame(bitcoin_price_data,columns=['TimeStamp','price'])
```

```python
data.head(10)
```

```python
data['Date']=pd.to_datetime(data['TimeStamp'],unit='ms')
```

```python
data.head(10)
```

```python
candlestick_data=data.groupby(data.Date.dt.date).agg({'price':['min','max','first','last']})
```

```python
!pip install plotly
import plotly.graph_objects as go
```

```python
fig=go.Figure(data=[go.Candlestick(x=candlestick_data.index,open=candlestick_data['price']['first'],high=candlestick_data['price']['max'], low=candlestick_data['price'],close=candlestick_data['price']['last'])])
```

```python
import plotly.offline as plotly
```

```python
plot=plotly.plot
```

```python
fig.update_layout(xaxis_rangeslider_visible=False, xaxis_title='Date',yaxis_title ='Price($USD)', title='Bitcoin Candlestick Chart')

plot(fig,filename='bitcoin_candlestick_graph.html')

```
